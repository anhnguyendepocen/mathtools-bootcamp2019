Now we are going to plot a tuning curve. What this means, is for each object (or location) we would like to calculate the mean number of spikes fired for that object across the whole trial. We then want to bar plot this for the 7 objects (and then the 3 locations). We will write a function to do this that takes in a label list and the data array and calculates a tuning curve for those labels

```python
def calc_tuning_curve(data,labels, label_names):
    tuning_curve_means = []
    tuning_curve_std = []
    for label in label_names:
        inds = np.where(labels == label)
        label_arr = data[inds]
        sum_spike = label_arr.sum(axis=1)
        tuning_curve_means.append(sum_spike.mean())
        tuning_curve_std.append(sum_spike.std())
    
    return [tuning_curve_means, tuning_curve_std]

tuning_curve_object = calc_tuning_curve(raster_array, object_labels, unique_objects)
tuning_curve_loc = calc_tuning_curve(raster_array, loc_labels, unique_locations)
print(tuning_curve_object, tuning_curve_loc)
```

```python
def plot_tuning_curve(tuning_curve, labels, title = None):
    y = tuning_curve[0]
    x = np.arange(len(y))
    error = tuning_curve[1]
    plt.figure()
    plt.errorbar(x,y, error)
    plt.title(title)
    plt.xticks(np.arange(len(y)), labels)
    plt.ylabel('Mean number of spikes fired to category')
    plt.xlabel('Category')
    plt.show()
    
plot_tuning_curve(tuning_curve_object, unique_objects, title = 'Object Tuning')
plot_tuning_curve(tuning_curve_loc, unique_locations, title = 'Location Tuning')
```

Now we are going to work with the data across all the objects. Our goal is to form a matrix of mean firing rates for the neurons for all object trials (i.e. 420 x num_neurons). However, as with any real data we need to do some cleaning first. 
1. An important quantity is the noise to signal ratio, known as the Fano Factor it is commonly thought that neurons follow a poisson distribution (i.e. mean = variance) so fano factor = 1. In real data; however, this is rarely true so we would like to only keep neurons that have a Fano factor lower than some threshold.

```python
#Now how do we compute the fano factor for each raster
def compute_fano(raster_data):
    mean_fr = np.mean(np.sum(raster_data, 1))
    var_fr = np.mean((np.sum(raster_data,1) - mean_fr)**2)
    
    fano_fact = var_fr/mean_fr
    
    return var_fr/mean_fr, mean_fr
```

```python
fano, _  = compute_fano(raster_array)
print(fano)
```

```python
#Compute fano factor across all neurons
import os
fano_facts = []
means = []
for f in os.listdir(raster_directory):
    raster_arr = loadmat(raster_directory + f)['rasterdata']
    fano, mean = compute_fano(raster_arr)
    fano_facts.append(fano)
    means.append(mean)

plt.scatter(means, fano_facts)
plt.axhline(y=1, color='r')
plt.title('Fano Factor vs. Mean')
plt.xlabel('Mean firing rate')
plt.ylabel('Fano Factor')

#We can see these neurons are not behaving like poisson and are infact generally exhibit very high trial-to-trial variability
```

Let's set a threshold at 8 and only keep neurons with a fano factor below 8. Now we want to loop again through the raster directory and create a matrix mean_dat that is 420 x num_clean neurons where each index of mean_dat[i,j]  has the mean firing rate of that neuron (after stimulus presentation - i.e. indexes 500:1000) across the trials

```python
#Since we don't know the output number of neurons- let's start with an empty array and append to it 
#We only add a neuron if the fano factor is below 8 and there are 420 trials

mean_object_dat = np.empty((420,0))

for f in os.listdir(raster_directory):
    raster_arr = loadmat(raster_directory + f)['rasterdata']
    fano, _ = compute_fano(raster_arr)
    mean_across_time = np.mean(raster_arr[:,500:1000], axis=1, keepdims=True)
    if fano < 8:
        mean_object_dat = np.append(mean_object_dat, mean_across_time, axis=1)
        

print(mean_object_dat.shape)
```

```python
#Are there patterns in mean firing?

unique_object_labels = ['car', 'couch', 'face', 'flower', 'guitar', 'hand', 'kiwi']
plt.imshow(mean_object_dat, aspect = 0.15)
plt.title('Mean Object Firing Rates')
plt.ylabel('Object category')
plt.xlabel('Neuron ID')
plt.yticks(np.arange(0, 420, 60), unique_object_labels, fontsize=8);
plt.colorbar()
```

It's hard to see any patterns here but some neurons have higher mean firing across all categories than others. How can we see if the neural responses are similar to each other for each object category.  We could look at correlation or covariance (brief explanation of difference). Correlation makes more sense here because we want it to be invariant to scaling of the magnitudes. 

```python
correlation = np.corrcoef(mean_object_dat)
plt.imshow(correlation)
plt.title('Correlation Between Object Categories')
plt.ylabel('Object Category')
plt.yticks(np.arange(0, 420, 60), unique_object_labels, fontsize=8);
plt.xticks([])
plt.colorbar()
```
